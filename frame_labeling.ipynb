```python
import pandas as pd
from PIL import Image
from skimage.restoration import estimate_sigma
import cv2 as cv
import numpy as np
import glob
```

```python
# yellow | 1 | down
# purple | 0 | up
max = 255
colors = {"yellow":(37, 255, 255),"purple":(160, 255, 128)}

def rgb_to_hsv(img):
    return cv.cvtColor(img, cv.COLOR_BGR2HSV)

def is_noisy(img, thresh = 0.15, v = False):
    inverse_noise = estimate_sigma(img, multichannel=True, average_sigmas=True)
    if v:
        print(f"inverse_noise: {inverse_noise}\n")
    return True if inverse_noise > thresh else False

def is_yellow(h):
    return np.abs(colors["yellow"][0] - h) < np.abs(colors["purple"][0] - h)

def find_color(img, noise_thresh = 0.16, sat_thresh = 0.35, v = False):
    if is_noisy(img, noise_thresh, v):
        return 2
    down_count = 0
    up_count = 0
    hsv_img = rgb_to_hsv(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            #print(hsv_img[i,j,:])
            sat = hsv_img[i,j,1]
            if sat > sat_thresh*255:
                if is_yellow(hsv_img[i,j,0]):
                    down_count+=1 
                else:
                    up_count+=1
    if v:
        print(f"up: {up_count}\ndown: {down_count}")
    return 1 if down_count > up_count else 0
```

```python
path = "input_transformed/RAFT_flow/pushup001/frame_0031.png"
img = cv.imread(path)
find_color(img, v=True)
#path = "validation_transformed/RAFT_flow/pushup001/frame_0272.png"
#img = cv.imread(path)
#print(is_noisy(img))
#path = "validation_transformed/RAFT_flow/pushup001/frame_0245.png"
#img = cv.imread(path)
#is_noisy(img)
```

```python
input_folders = glob.glob('input_transformed/RAFT_flow/*')
df = pd.DataFrame()
i = 0
for input_path in input_folders:
    input_path = input_path.replace("\\","/")
    input_folder = glob.glob(f'{input_path}/*')
    print("*")
    for input in input_folder:
        df.loc[i,0] = input
        img = cv.imread(input)
        df.loc[i,1] = find_color(img)
        print(df.loc[i,1])
        i+=1
df.to_csv('input_labels.csv', header = False, index = False)
```
