```python
import cv2 as cv
import numpy as np
import glob
import matplotlib.pyplot as plt
import re
import os
```

```python
def convertVid(input_path, frame_div = 3, img_size = (128, 128), v=False):

    cap = cv.VideoCapture(input_path)
    
    i = 0
    video_length = int(cap.get(cv.CAP_PROP_FRAME_COUNT))-1
    if v:
        print(f"frame count: {video_length}")
    
    while(cap.isOpened()):
        try:
            ret, frame = cap.read()
            if not ret:
                continue
            frame = cv.resize(frame, img_size)

            if i % frame_div == 0:
                cv.imwrite(f"input_transformed/vid/pushup{int(input_path[12:-4]):03d}/frame_{(i//frame_div):04d}.png", frame)
            i+=1
            if i > video_length-1:
                cap.release()
                break
        except Exception as e:
            print(f'error: {e}')
    if v:
        print(f"{i//frame_div} frames were captured")

```

```python
# Gather input files
if not os.path.exists("input_transformed/vid/"):
    os.mkdir("input_transformed/vid/")
inputs = glob.glob('input/*')

for input_path in inputs:
    input_path = input_path.replace("\\","/")
    if not os.path.exists(f"input_transformed/vid/pushup{int(input_path[12:-4]):03d}/"):
        os.mkdir(f"input_transformed/vid/pushup{int(input_path[12:-4]):03d}/")
    convertVid(input_path, 3, (128,128), True)
    
```
